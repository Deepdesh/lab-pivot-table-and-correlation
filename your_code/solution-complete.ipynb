#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python

pivot5 = employee.pivot_table(index=['Title','Department'],values=['Salary'],aggfunc =['mean','median','max','min','std'],dropna=False)

pivot5
```

**Finding 1:** 
There are NaN values in a standard deviation part, this is resulting for the titles in each departments who only have 1 count. This makes sense as pivot table cannot calculate deviation/variance to the mean. In these cases, we can replace NaN with 0 as there are no deviations.


#### Based on your comments, fill the missing values with an appropriate value.

```python
# replacing NaN with 0

pivot5 = employee.pivot_table(index=['Title','Department'],values='Salary',aggfunc = [np.mean, np.median,np.max,np.min,np.std],dropna = False,fill_value=0)

pivot5

```

**finding 2:** Sales has mean=max=median, and the count is two. This could be problematic later, if we decided to drop max values.

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
#finding values for max salaries across departments to visualize
pivot8 = employee.pivot_table(index=['Department'],values=['Salary'],aggfunc = {'Salary':np.max})
pivot8

employee.head()
```

```python
#define function: remove_max()

def remove_max(row):
    
    if (row['Department']=='IT') & (row['Salary']==70):
        return None
    elif (row['Department']=='HR') & (row['Salary']==70):
        return None
    else:
        return row['Salary']
#Note based on finding 2: Decided not to drop the max salary of Sales department because there are only two salaries value,
#and they are equal to each other, dropping them could give rise to misleading inferences. salaries for sales would be 0 then.

employee_c = employee
#creating employee_c = copy of dataframe employee
```

```python
#applying to column salary in dataframe copy
employee_c['Salary'] = employee.apply(remove_max, axis=1)
pivot_final = employee_c.pivot_table(index= ['Department'], values= ['Salary'] , aggfunc= [np.mean])
pivot_final
```

If I had dropped sales max values, the mean salary of the Sales department would not appear here as the values would be NaN.
